```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import rc
```

```python
data = np.genfromtxt("masa_l.txt", delimiter = '\t')
x = np.linspace(0, 95, len(data))
```

```python
x
```

```python
f = lambda x,a,b: a*x+b
popt, pcov = curve_fit(f, np.log(x[1:]), np.log(data[1:]), p0=[-1.8, 0.0])
print(popt, "\n", pcov)
```

```python
plt.plot(np.log(x[1:]), np.log(data[1:]), 'r.', label = 'Datos')
plt.plot(np.log(x), f(np.log(x), popt[0], popt[1]), 'k-', label = 'Ajuste')
plt.text(2.5, -1.35, r'$D-2 = %.5f \pm %.5f$'%(popt[0],pcov[0,0]), fontsize = 15)
plt.title(r'log densidad del cluster percolante vs $l$ para $L = 64$')
plt.xlabel(r'log $l$')
plt.ylabel(r'log $\rho$')
plt.grid()
plt.legend(loc = 'best')
#plt.yscale('log')
#plt.xscale('log')
plt.show()
#print(np.log(data[10])-np.log(data[5])/(np.log(x[10])-np.log(x[5])))
```
