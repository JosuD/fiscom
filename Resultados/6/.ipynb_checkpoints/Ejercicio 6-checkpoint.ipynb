```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rc
from scipy.optimize import curve_fit
from ipywidgets import widgets as wid
```

```python
data = np.genfromtxt('m2.txt', delimiter = '\t')
p = np.linspace(0.45, 0.65, np.shape(data)[0])
dif = p[6]-p[5]
plt.plot(p,data, '.', p[68], data[68], 'ro')
plt.yscale('log')
plt.xscale('log')
plt.show()
```

```python
ldata = np.log(data)
lp = np.log(p)
ldif = np.log(dif)
redot = 61
posdot = 70
g = lambda x, a, b: a*x+b


largo = 13
#for largo in range(5,10):
lp_p = lp[posdot:posdot+largo]
ldata_p = ldata[posdot:posdot+largo]
lp_m = lp[redot-largo:redot]
ldata_m = ldata[redot-largo:redot]
popt1, pcov1 = curve_fit(g, lp_p, ldata_p)
popt2, pcov2 = curve_fit(g, lp_m, ldata_m)


plt.plot(lp, ldata, 'b.', lp[redot], ldata[redot], 'ro', lp[posdot], ldata[posdot], 'ro')
plt.plot(lp_p, ldata_p, 'g.', lp_m, ldata_m, 'k.')
plt.plot(lp_p, g(lp_p, popt1[0], popt1[1]),'k-', lp_m, g(lp_m, popt2[0], popt2[1]),'k-' )
plt.show()
plt.grid(True)
print(popt1[0], popt2[0])
```

```python
g = lambda x, a, b: a*x+b
predot = 60
posdot = 70
gamma_p = np.zeros((25))
gamma_m = np.zeros((25))
i=0
for largo in range(5, 30):
    lp_p = lp[posdot:posdot+largo]
    ldata_p = ldata[posdot:posdot+largo]
    lp_m = lp[predot-largo:predot]
    ldata_m = ldata[predot-largo:predot]
    popt1, pcov1 = curve_fit(g, lp_p, ldata_p)
    popt2, pcov2 = curve_fit(g, lp_m, ldata_m)
    gamma_p[i] = popt1[0]
    gamma_m[i] = popt2[0]

    i += 1
largo = range(5,30)
pc = 5.939410454074074419e-01
plt.plot(np.fabs(p[5:30]-pc), -gamma_p,'r.', label= r'$\gamma^+$')
plt.plot(np.fabs(p[5:30]-pc), gamma_m, 'b.', label = r'$\gamma^-$')
plt.grid(True)
plt.legend()
plt.xlabel(r'$|p-p_c|$')
plt.ylabel(r'$\gamma$')
plt.title('Gamma matching')
plt.text(0.11, 20, r'$\gamma = %.2f$'%(gamma_m[8]), fontsize = 16)
plt.savefig('matching.png', format = 'png', dpi = 200)

plt.show()

```
