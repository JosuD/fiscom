```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rc
from scipy.optimize import curve_fit
import os
```

```python
a1 = 'tamanios_s1.txt'
a2 = 'tamanios_s2.txt'
a3 = 'tamanios_s3.txt'
a4 = 'tamanios_s4.txt'
a5 = 'tamanios_s5.txt'
a6 = 'tamanios_s6.txt'
a7 = 'tamanios_s7.txt'
a8 = 'tamanios_s8.txt'
ns = np.zeros((100,8))
ns[:,0] = np.genfromtxt(a1, delimiter = '\t')
ns[:,1]= np.genfromtxt(a2, delimiter = '\t')
ns[:,2]= np.genfromtxt(a3, delimiter = '\t')
ns[:,3]= np.genfromtxt(a4, delimiter = '\t')
ns[:,4]= np.genfromtxt(a5, delimiter = '\t')
ns[:,5]= np.genfromtxt(a6, delimiter = '\t')
ns[:,6]= np.genfromtxt(a7, delimiter = '\t')
ns[:,7]= np.genfromtxt(a8, delimiter = '\t')
p = np.linspace(0.45, 0.65, len(ns[:,0]))
```

```python
ns = np.zeros((100,8))
np.shape(ns)
```

```python
plt.plot(ns[:,0], '.')
plt.plot(ns[:,4], '.')
plt.plot(ns[:,7], '.')
plt.show()
print(len(ns[:,0]))
```

```python
indice = np.zeros((8))
s = np.zeros((8))
s0 = 5
pc = 5.925807619259259739e-01
for i in range(8):
    lista = ns[:,i].tolist()
    indice[i] = lista.index(max(lista))
    s[i] = s0 + 2*i

f = lambda x, a, b: a*x+b
popt,pcov = curve_fit(f, np.log(s), np.log(np.fabs(p[indice.tolist()]-pc)))
print(popt, '\n', pcov)
print(p[indice.tolist()])
plt.plot(np.log(s),np.log(np.fabs(p[indice.tolist()]-pc)),'.')
plt.plot(np.log(s), f(np.log(s), popt[0], popt[1]), 'r-', label = 'Ajuste')
plt.ylabel(r'$log |p_{max}-p_c|$')
plt.xlabel(r'$log{s}$')
plt.title(r'Ajuste para hallar $\sigma = %.4f \pm %.4f$'%(popt[0], pcov[0,0]))
plt.legend()
plt.grid(True)
plt.savefig('pmaxvss.png', format = 'png', dpi = 200)

plt.show()
```
